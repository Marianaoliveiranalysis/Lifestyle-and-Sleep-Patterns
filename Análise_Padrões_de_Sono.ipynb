```python
import pandas as pd
import numpy as np
```

```python
# Carregar a base de dados
df = pd.read_csv('/content/Sleep_health_and_lifestyle_dataset.csv')
df.head()
```

```python
#Diagnóstico da base de dados
df.info()
df.describe(include='all')
```

```python
# Verificação de valores ausentes e duplicatas
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
# Limpeza e Padronização dos dados
df['Sleep Disorder'] = df['Sleep Disorder'].fillna('Desconhecido')
df.head()
```

```python
df.isnull().sum()
```

```python
# Criação de colunas derivadas
# 1. Criar coluna categorizando a Duração do Sono
def categorizar_sono(horas):
    if horas < 6:
        return "Curto"
    elif 6 <= horas <= 8:
        return "Ideal"
    else:
        return "Longo"

df["Sleep Category"] = df["Sleep Duration"].apply(categorizar_sono)
df.head()
```

```python
# 2. Classificação de Bem-estar do Sono com np.where - Numpy
df["Sleep Wellness"] = np.where(
    (df["Quality of Sleep"] >= 7) & (df["Stress Level"] <= 4), "Bom",
    np.where(
        (df["Quality of Sleep"] >= 5) & (df["Stress Level"] <= 6), "Regular",
        "Ruim"
    )
)
df.head()
```

```python
 # Análise de Tendências Centrais

# Média
media = df.mean(numeric_only=True).to_frame(name="Média")
media.head()
```

```python
# Mediana
mediana = df.median(numeric_only=True).to_frame(name="Mediana")
mediana.head()
```

```python
# Moda (pegando a primeira moda de cada coluna)
moda = df.mode().iloc[0].to_frame(name="Moda")
moda.head()
```

```python
# Medidas de dispersão
# Variância
variancia = df.var(numeric_only=True).to_frame(name="Variância")
variancia.head()
```

```python
# Valores mínimos
minimos = df.min(numeric_only=True).to_frame(name="Mínimo")
minimos.head()
```

```python
# Valores máximos
maximos = df.max(numeric_only=True).to_frame(name="Máximo")
maximos.head()
```

```python
desvio_padrao = df.std(numeric_only=True).to_frame(name="Desvio Padrão")
desvio_padrao.head()
```

```python
# Salvamento do arquivo tratado
df.head()
df.to_csv('Sleep_health_and_lifestyle_dataset_final_Tratado.csv', index=False)
```
